```python
# 3. Explain the following terms:
#  - The differnce in np.array, np.asarray, np.asanyarray
#  - The differnce between deep copy and shallow copy


a = [1,2,3]
b = np.array(a) # Always creates a new array
print(b)

a = [1, 2, 3]
b = np.asarray(a)  # No copy if 'a' is already a NumPy array
print(b)

a = np.matrix([[1,2],[3,4]])
b = np.asanyarray(a) #'b' retains the matrix subclass
print(b)

import copy

original_list = [1, [2, 3], 4]
deep_copied_list = copy.deepcopy(original_list)
deep_copied_list[1][0] = 99

print("\nOriginal list:", original_list)
print("Deep copied list:", deep_copied_list)

original_list = [1, [2, 3], 4]
shallow_copied_list = copy.copy(original_list)
shallow_copied_list[1][0] = 99

print("\nOriginal list:", original_list)
print("Shallow copied list:", shallow_copied_list)
```

```python
# 5. Create a NumPy array with random integers between 1 and 10 of shape (5, 6). After creating the array perform the following operations:

#a)Extract all even integers from array.
#b)Extract all odd integers from array.

random_array = np.random.randint(1,11, (5,6))
print("Original array:\n", random_array)

even_integers = random_array[random_array % 2 == 0]
print("\nEven integers:\n",even_integers)

odd_integers = random_array[random_array % 2 != 0]
print("\nOdd integers:\n",odd_integers)
```

```python
# 6. Create a 3D NumPy array of shape (3, 3, 3) containing random integers between 1 and 10. Perform the following operations:

#a) Find the indices of the maximum values along each depth level (third axis).
#b) Perform element-wise multiplication of between both array

array_3d = np.random.randint(1, 11, (3, 3, 3))
print("Original 3D array:\n", array_3d)

max_indices = np.argmax(array_3d, axis=2)
print("\nIndices of maximum values along each depth level:\n", max_indices)

array_3d_2 = np.random.randint(1, 11, (3, 3, 3))
print("\nSecond 3D array:\n", array_3d_2)

result_array = array_3d * array_3d_2
print("\nResult of element-wise multiplication:\n", result_array)
```

```python
# 15.Create a NumPy array data containing 1000 samples from a normal distribution. Perform the following tasks using Matplotlib:
#a) Plot a histogram of the data with 30 bins.
#b) Overlay a line plot representing the normal distribution's probability density function (PDF).
#c) Label the x-axis as 'Value' and the y-axis as 'Frequency/Probability'.
#d) Set the title of the plot as 'Histogram with PDF Overlay'.

from scipy.stats import norm
data = np.random.randn(1000)

plt.hist(data, bins=30, density=True, alpha=0.6, color='g', label='Histogram')
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, np.mean(data), np.std(data))
plt.plot(x, p, 'k', linewidth=2, label='PDF')

plt.xlabel('Value')
plt.ylabel('Frequency/Probability')
plt.title('Histogram with PDF Overlay')
plt.legend()
plt.show()
```
